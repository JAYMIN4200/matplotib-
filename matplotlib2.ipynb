```python
#generate some data
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x,y)

plt.xlabel("category")
plt.ylabel('value')
plt.title('bar chat ex')

plt.show()
```

```python
categories=np.array(['A','B','C','D'])
v1=np.array([3,8,1,10])
v2=np.array([6,2,7,5])

bar_width=0.45

plt.bar(categories,v1,bar_width,label='dataset1')
plt.bar(categories,v2,bar_width+0.25,label='dataset2')

plt.xlabel('categories')
plt.ylabel('value')
plt.title('grp barchat ex')

plt.legend()
plt.show()
```

```python
categories=np.array(['A','B','C','D'])
v1=np.array([3,8,1,10])
v2=np.array([6,2,7,5])
v3=np.array([5,13,12,2])
bar_width=0.25

plt.bar(categories,v1,bar_width,label='dataset1')
plt.bar(categories,v2,bar_width+0.10,label='dataset2')
plt.bar(categories,v3,bar_width+0.1,label='dataset3')

plt.xlabel('categories')
plt.ylabel('value')

plt.title('grp barchat ex')

plt.legend()
plt.show()
```

```python
categories=np.array(['A','B','C','D'])
v1=np.array([3,8,1,10])

e=(0,0,0.1,0)
plt.pie(v1,labels=categories,explode=e,autopct='%1.1f%%')



plt.xlabel('categories')
plt.ylabel('value')
plt.title('pie chat ex')


plt.show()
```
